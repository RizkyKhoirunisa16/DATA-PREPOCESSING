```python
import spacy
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import nltk

# Unduh data 'stopwords' NLTK
nltk.download('stopwords')

# Muat model bahasa
nlp = spacy.load("xx_ent_wiki_sm")

# Kalimat asli
k = "JKN-KIS menanggung perawatan penyakit Orang Dengan Gangguan Jiwa, agar tidak tercipta Joker lainnya :)"
print('Kalimat asli')
print(k)
print()

# Tahapan Preprocessing

# Case Folding (menjadi huruf kecil)
case_folding = k.lower()
print("Case Folding: ")
print(case_folding)
print()

# Filtering (menghapus tanda baca)
fil = case_folding.translate(str.maketrans("", "", string.punctuation))
print("Filtering")
print(fil)
print()

# Stemming (mengubah kata dasar)
print('Stemming')
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stemm = stemmer.stem(fil)
print(stemm)
print()

# Tokenize (menggunakan spaCy untuk tokenisasi)
doc = nlp(stemm)
tokens = [token.text for token in doc]
print("Tokenize")
print(tokens)
print()

# Stopword Removal (menghapus kata yang tidak berarti)
from nltk.corpus import stopwords
stopword = set(stopwords.words('indonesian'))
removed = []
for t in tokens:
    if t not in stopword:
        removed.append(t)
print('Stopwords Removal')
print(removed)
```

```python
import requests
from bs4 import BeautifulSoup
import string
import nltk
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

nltk.download('punkt_tab')
```

```python
#URL Berita
url = 'https://www.cnnindonesia.com/hiburan/20250506223519-226-1226502/ryan-adriandhy-dan-kesuksesan-jumbo-yang-beyond-expectation'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

#Ambil paragraf pertama yang bukan atau kosong
paragraf = soup.find_all('p')
for p in paragraf:
  if p.get_text().strip():
    original_text = p.get_text().strip()
    break

print("Paragraf Pertama:\n", original_text)
```

```python
case_folding = original_text.lower()
print("\n Case Folding:\n", case_folding)
```

```python
filtered = case_folding.translate(str.maketrans("", "", string.punctuation))
print("\n Filtering:\n", filtered)
```

```python
tokens = word_tokenize(filtered)
print("\n Tokenize:\n", tokens)
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

stemmed = [stemmer.stem(word) for word in tokens]
print("\n Stemming:\n", stemmed)
```
